```python
city_mapping = {
    'Aberdeen': 'Aberdeen City',
    'Glasgow': 'Glasgow City',
    'Edinburgh': 'City of Edinburgh',
    'Orkney': 'Orkney Islands',
    'Shetland': 'Shetland Islands',
    'Dundee': 'Dundee City',
    'Western Isles - Eilean Siar': 'Na h-Eileanan Siar'
}

df['source_location'] = df['source_location'].replace(city_mapping)

df['source_location'].unique()
```

```python
import pandas as pd
import numpy as np
from calendar import monthrange

# Example: your original dataframe
# df = pd.DataFrame({
#     'year': [2023, 2023],
#     'month': [1, 1],
#     'source_location': ['A', 'B'],
#     'detected_crime': [93, 120]
# })

# Create an empty list to store the rows of the new dataframe
rows = []

# Iterate through each row in the original df
for _, row in df.iterrows():
    year = row['year']
    month = row['month']
    source = row['source_location']
    total_crime = row['DETECTED CRIME']
    
    # Get number of days in the month
    num_days = monthrange(year, month)[1]
    
    # Random distribution base: start with equal distribution
    base = total_crime // num_days
    crimes = [base] * num_days

    # Distribute the remainder randomly with small adjustments
    remainder = total_crime - sum(crimes)
    for _ in range(remainder):
        index = np.random.randint(0, num_days)
        crimes[index] += 1

    # Optionally, add small noise to the base (not affecting sum)
    noise = np.random.randint(-1, 2, size=num_days)
    crimes = np.array(crimes) + noise

    # Ensure no value goes below 0
    crimes = np.maximum(crimes, 0)

    # Calculate the difference after noise
    total_after_noise = crimes.sum()
    diff = total_crime - total_after_noise

    # If there is a difference, adjust to balance the total
    if diff != 0:
        adjust_indices = np.random.choice(num_days, abs(diff), replace=True)
        for idx in adjust_indices:
            crimes[idx] += 1 if diff > 0 else -1

    # Ensure no value goes below 0 after adjustment
    crimes = np.maximum(crimes, 0)

    # Recalculate the total after adjustment to ensure it equals the original total
    total_after_adjustment = crimes.sum()

    # If the sum is still not equal to total_crime, adjust the last crime entry
    if total_after_adjustment != total_crime:
        diff = total_crime - total_after_adjustment
        crimes[-1] += diff

    # Append to rows
    for day, daily_crime in enumerate(crimes, 1):
        rows.append({
            'year': year,
            'month': month,
            'day': day,
            'source_location': source,
            'DETECTED CRIME': daily_crime
        })

# Create the new dataframe
ndf = pd.DataFrame(rows)
```
